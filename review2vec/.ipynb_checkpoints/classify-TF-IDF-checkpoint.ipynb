```python
import jieba
import json
import os
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
```

```python
directory = "ÁîúÈªû"
file_names = os.listdir(directory)
```

```python
all_comments = []

# ÈÅçÊ≠∑ÁõÆÈåÑ‰∏≠ÁöÑÊâÄÊúâÊñá‰ª∂
file_names = os.listdir(directory)
for file_name in file_names:
    file_path = os.path.join(directory, file_name)
    if not os.path.isfile(file_path): continue
    if not file_path.endswith('.json'): continue
    if os.path.getsize(file_path) == 0: continue
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)
            comments = data.get('comments', [])
            all_comments.extend(comments)
    except: 
        print("something wrong in "+file_path)
```

```python
df = pd.DataFrame(all_comments)
df = df.drop(columns='other').dropna(subset=['comment'])

# Â∞çË©ïË´ñÈÄ≤Ë°åÂàÜË©ûÔºåÂÖàÈÄ≤Ë°åÈ°ûÂûãÊ™¢Êü•ÂíåËôïÁêÜ
df['comment'] = df['comment'].astype(str).apply(lambda x: ' '.join(jieba.cut(x)))
```

```python
df
```

```python
# ‰ΩøÁî® TF-IDF ÈÄ≤Ë°åÁâπÂæµÊèêÂèñ
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['comment'])
```

```python
# ‰ΩøÁî®ËÇòÈÉ®Ê≥ïÂâáÁ¢∫ÂÆöÊúÄ‰Ω≥ÁöÑËÅöÈ°ûÊï∏Èáè
sse = []
for k in range(1, 21):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

# Áπ™Ë£ΩËÇòÈÉ®Ê≥ïÂâáÂúñ
fig = plt.figure(figsize=(10, 7))
plt.plot(range(1, 21), sse, marker='o')
plt.title('Elbow Method for Optimal n_clusters')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
fig.savefig("elbow.jpg")
```

```python
# ‰ΩøÁî® K-means ÈÄ≤Ë°åËÅöÈ°û
kmeans = KMeans(n_clusters=4, random_state=23)
df['cluster'] = kmeans.fit_predict(X)
```

```python
# ‰ΩøÁî® TSNE Êàñ PCA ÈÄ≤Ë°åÈôçÁ∂≠‰ª•‰æøÂèØË¶ñÂåñ
educed_data = PCA(n_components=2).fit_transform(X.toarray())
```

```python
sil_score = silhouette_score(X, df['cluster'])
print(f"Silhouette Score: {sil_score}")
```

```python
# Áπ™Ë£ΩÁµêÊûú
fig = plt.figure(figsize=(10, 7))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=df['cluster'], cmap='jet', s=0.1)
plt.title('K-means Clustering of Reviews')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# fig.savefig("K-means Clustering of Reviews.jpg")
```

```python
# ‰ΩøÁî®Isolation ForestÈÄ≤Ë°åÁï∞Â∏∏Ê™¢Ê∏¨
isolation_forest = IsolationForest(contamination=0.1, random_state=42)
df['anomaly'] = isolation_forest.fit_predict(X)# ‰ΩøÁî®Isolation ForestÈÄ≤Ë°åÁï∞Â∏∏Ê™¢Ê∏¨
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=df['anomaly'], cmap='coolwarm', s=0.1)
plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar(label='Anomaly')
plt.show()
```
